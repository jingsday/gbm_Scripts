```python
adata_list = []

# Loop over each sample and read in the AnnData object
for name in names_list:

    mtx =f"{name}_matrix.mtx.gz"
    adata = sc.read_mtx(mtx)
    cells=pd.read_csv(f'{name}_barcodes.tsv.gz',header=None)
    features=pd.read_csv(f'{name}_features.tsv.gz',header=None,sep='\t')
    adata= adata.T
    #check the columns first to make sure they are the ones you need 
    adata.obs['CellID']= cells[0].tolist()
    adata.obs.index = adata.obs['CellID']
    adata.var['Gene']= features[0].tolist()
    adata.var.index= adata.var['Gene']
    adata.var_names_make_unique() 
    adata.var['mt'] =adata.var_names.str.startswith('MT-')


    sc.pp.filter_cells(adata, min_genes=200)
    sc.pp.calculate_qc_metrics(adata,qc_vars=['mt'], percent_top=None, log1p=False, inplace=True)
    adata= adata[adata.obs.n_genes_by_counts <6000, :]
    adata= adata[adata.obs.pct_counts_mt< 5, :].copy()


    adata.obs['source'] = name[11:]
    adata.layers["counts"] = adata.X.copy()    
    sc.pp.normalize_total(adata, target_sum=1e4)
    sc.pp.log1p(adata)

    print(adata)
    adata.raw = adata  # keep full dimension safe
    adata_list.append(adata)
```

```python

# Extract the result dictionary
result = adata_singlet.uns["rank_genes_groups"]

# Get list of group names (e.g., clusters)
groups = result["names"].dtype.names

# Store filtered results here
filtered_dfs = []

# Loop over each group and extract + filter
for group in groups:
    df = pd.DataFrame({
        "gene": result["names"][group],
        "log2fc": result["logfoldchanges"][group],
        "pvals": result["pvals"][group]
    })

    # Filter by thresholds
    #df_filtered = df[(abs(df["log2fc"]) > 1) & (df["pvals"] < 0.05)].copy()
    df_filtered = df[df["pvals"] < 0.05].copy()

    df_filtered["group"] = group
    filtered_dfs.append(df_filtered)

# Combine all filtered results into one DataFrame
filtered_genes_all = pd.concat(filtered_dfs, ignore_index=True)

# Show top 5
filtered_genes_all.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from adjustText import adjust_text

plt.figure(figsize=(10, 7))

# Scatterplot
sns.scatterplot(
    data=filtered_genes_all,
    x='log2fc',
    y='-log10(p-value)',
    hue='color',
    palette={'red': 'red', 'blue': 'blue', 'grey': 'grey'},
    alpha=0.9,
    edgecolor=None,
    legend=False
)

# Threshold lines
plt.axhline(-np.log10(0.05), color='gray', linestyle='--', linewidth=1)
plt.axvline(1, color='gray', linestyle='--', linewidth=1)
plt.axvline(-1, color='gray', linestyle='--', linewidth=1)

# Top genes to annotate
top_pos = filtered_genes_all[filtered_genes_all['gene'].isin([
    'XIST', 'GPR160','AC004603.4', 'ADAMTS18', 'CARD14', 'CTNNA3',
    'NEAT1', 'COL4A5', 'TMEM144', 'PEX5L'
])]

top_neg = filtered_genes_all[filtered_genes_all['log2fc'] < -1].nsmallest(5, 'log2fc')
highlight_genes = pd.concat([top_pos, top_neg])

# Annotate genes
texts = []
for _, row in highlight_genes.iterrows():
    texts.append(
        plt.text(
            row['log2fc'],
            row['-log10(p-value)'],
            row['gene'],
            ha='center',
            va='center'
        )
    )

# Adjust to avoid overlap with fake arrows (just lines)
adjust_text(
    texts,
    expand_text=(1.1, 1.3),
    expand_points=(1.1, 1.3),
    arrowprops=dict(
        arrowstyle='-',  # simple line (may be invisible depending on backend)
        color='black',
        lw=0.5
    ),
    force_text=0.5,
    force_points=0.3,
    only_move={'points': 'y', 'text': 'xy'}
)

plt.xlabel('Fold Change(Log2)')
plt.ylabel('Significance(-log10)')
plt.tight_layout()
plt.show()
```

```python


def palplot(pal, names, colors=None, size=1):
    n = len(pal)
    f, ax = plt.subplots(1, 1, figsize=(n * size, size))
    ax.imshow(np.arange(n).reshape(1, n),
              cmap=mpl.colors.ListedColormap(list(pal)),
              interpolation="nearest", aspect="auto")
    ax.set_xticks(np.arange(n) - .5)
    ax.set_yticks([-.5, .5])
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    colors = n * ['k'] if colors is None else colors
    for idx, (name, color) in enumerate(zip(names, colors)):
        ax.text(0.0+idx, 0.0, name, color=color, horizontalalignment='center', verticalalignment='center')
    return f

```
